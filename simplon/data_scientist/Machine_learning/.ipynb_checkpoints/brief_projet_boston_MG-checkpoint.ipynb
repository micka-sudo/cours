### Brief projet Boston

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
import seaborn as sns
```

### Chargement du data set boston

```python
dataset = datasets.load_boston()
```

### Transformer le jeu de données en data frame


```python
# pour avoir le nom des colonnes
df_boston_X = pd.DataFrame(dataset.data, columns=dataset.feature_names)
print(df_boston_X.shape)
# creation variable y liée au dataset
df_boston_y = dataset.target
# df_boston_y = df_boston_y.reshape(df_boston_y.shape[0], 1)
print(df_boston_y.shape)
# liée colonne du df X prix, avec le Df y
df_boston_X["Prix"] = df_boston_y

```

### Supprimer les valeurs nulles, s’il en existe.

```python
# Valeurs manquantes par colonnes
num_missing_X = lambda x: sum(x.isnull())
print("Valeurs manquantes par colonne:\n", df_boston_X.apply(num_missing_X, axis=0))

valeur_null_X = df_boston_X.index[df_boston_X.isnull().any(axis=1)]
valeur_null_X.shape
# Valeurs manquantes par colonnes
num_missing_y = lambda x: sum(x.isnull())
print("Valeurs manquantes par colonne:\n", df_boston_y.apply(num_missing_y, axis=0))
valeur_null_y = df_boston_y.index[df_boston_y.isnull().any(axis=1)]
valeur_null_y.shape
# pas de valeur null
```

### Choisir les variables explicatives (features) en menant une analyse de corrélation.


```python

matrice_corr = df_boston_X.corr().round(1)
sns.heatmap(data=matrice_corr,annot=True)
```

### Pour effectuer division dataset en training et test à 20% du df test_size =0.2 et utilisation d'une feature

```python
from sklearn.model_selection import train_test_split
x_train , x_test, y_train, y_test = train_test_split(df_boston_X,df_boston_y,test_size = 0.2, random_state=5)
print(x_test.shape)
y_test = y_test.reshape(y_test.shape[0], 2)
print(y_test.shape)
```

### Creation d'un objet de creation linéaire

```python
regr = linear_model.LinearRegression()
```

### Entrainement du modef utilsant l'ensemble entrainement

```python
regr.fit(x_train, y_train)
```

### Faire des prediction utilisant l'ensemble test

```python
boston_y_pred = regr.predict(x_test)
```

### Coefficients

```python
print('Coefficients: \n', regr.coef_)
```

### Erreur quadratique moyenne

```python
print('Mean squared error: %.2f'
      % mean_squared_error(y_test, boston_y_pred))
```

### Coefficient de determination: 1 = perfect prediction

```python
print('Coefficient of determination: %.2f'
      % r2_score(y_test, boston_y_pred))
```

### Graphique

```python
plt.scatter(x_test, y_test,  color='black')
plt.plot(x_test, boston_y_pred, color='blue', linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```
