# Arbre de régression
## Dataset : Boston

## librairies

```python
from sklearn.datasets import load_boston
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
from sklearn import model_selection
from sklearn.model_selection import train_test_split,ShuffleSplit
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, make_scorer, SCORERS
import numpy as np
import pandas as pd
from IPython.display import Image  
from sklearn.externals.six import StringIO  
import pydot  
from sklearn import tree
```

### Importation des données

```python
x = load_boston()
df = pd.DataFrame(x.data, columns = x.feature_names)
df["MEDV"] = x.target
X = df.drop("MEDV",1)   #Feature Matrix
y = df["MEDV"]          #Target Variable
df
```

### Scoring function

```python
def rmse(y,y_pred):
    return np.sqrt(mean_squared_error(y, y_pred))
```

### Best estimator 

```python
def bestRegressor(X,y,param_grid={'criterion':['mse']},scoring=make_scorer(rmse,greater_is_better=False),cv=10) : 
    grid_search = GridSearchCV(estimator=DecisionTreeRegressor(), param_grid=param_grid, scoring=scoring, cv=cv)
    grid_search.fit(X,y)
    return grid_search.best_estimator_, grid_search.best_params_,grid_search.best_score_,grid_search.cv_results_
```

```python
best = bestRegressor(X,y,param_grid={'max_depth':np.arange(1,10),'max_features':np.arange(2,8)},scoring='r2',cv=ShuffleSplit(n_splits = 10, test_size = 0.20, random_state = 0))
```

```python
print('Meilleur modèle :')
print(best[0])
print('meilleurs paramètres :')
print(best[1])
print('meilleur score :')
print(best[2])
print('résultat :')
print(best[3])
```

## Decision Tree Regression Visualization 

```python
dot_data = StringIO()  
tree.export_graphviz(best[0], out_file=dot_data,  
    filled=True, rounded=True,
    feature_names=x.feature_names,
    special_characters=True)
graph = pydot.graph_from_dot_data(dot_data.getvalue())  
Image(graph[0].create_png())  
```
