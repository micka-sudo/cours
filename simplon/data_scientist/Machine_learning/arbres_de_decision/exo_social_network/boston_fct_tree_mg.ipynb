### Importation librarie

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

### imporation base

```python
boston = load_boston()
X = boston.data
y= boston.target
```

### foncton arbre de decision

```python
def decisionTreeRgression(X, y, max_depth= None, split_param={'test_size':0.2,'random_state':0}):
    #split set
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = split_param['test_size'], random_state = split_param['random_state'])
    # training reg 
    clf = tree.DecisionTreeRegressor(max_depth=max_depth)
    clf = clf.fit(X_train, y_train)
    # graph tree_reg
    tree.plot_tree(clf, filled=True, fontsize=6)
    # predcition test
    y_pred = clf.predict(X_test)
    # metric model
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    adj_r2 =1-(1-r2)*(X_test.shape[0]-1)/(X_test.shape[0]-X_test.shape[1]-1)
    print('MAE : ',mae)
    print('MSE : ',mse)
    print('R² : ',r2)
    print('Adjusted R² : ',adj_r2)
    return [mae,mse,rmse,r2,adj_r2]
```

```python
regressor = decisionTreeRgression(X,y,split_param={'test_size':0.2,'random_state':0})
```

```python
regressor = decisionTreeRgression(X,y,max_depth=2,split_param={'test_size':0.3,'random_state':0})
```
