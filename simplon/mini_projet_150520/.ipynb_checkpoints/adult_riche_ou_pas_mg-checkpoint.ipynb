#### gestion des valeur manquantes avec SimpleInputer
- Remplacement par le "most-frequent"

```python
adult_df[['workclass', 'education', 'marital.status','occupation','relationship','race','sex','native.country','income']].apply(lambda x: x.astype('category'))
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(missing_values = np.nan, strategy="most_frequent")
adult_df_category = adult_df[['workclass','occupation','native.country']]
imputer = imputer.fit(adult_df_category[['workclass','occupation','native.country']])
adult_df_category = imputer.transform(adult_df_category[['workclass','occupation','native.country']])
adult_df_category = pd.DataFrame(data=adult_df_category , columns=[['workclass','occupation','native.country']])
adult_df_category.head()
```

#### remplacement valeur manquante par la moyenne

```python
imputer1 = SimpleImputer(missing_values = np.nan, strategy="mean")
adult_df_numeric = adult_df[['age','fnlwgt','education.num','capital.gain','capital.loss','hours.per.week']]
imputer1 = imputer1.fit(adult_df_numeric[['age','fnlwgt','education.num','capital.gain','capital.loss','hours.per.week']])
adult_df_numeric = imputer1.transform(adult_df_numeric[['age','fnlwgt','education.num','capital.gain','capital.loss','hours.per.week']])
adult_df_numeric = pd.DataFrame(data=adult_df_numeric , columns=[['age','fnlwgt','education.num','capital.gain','capital.loss','hours.per.week']])
```

#### Fusion avec le dataset d'origine

```python
adult_df[['workclass','occupation','native.country']] = adult_df_category[['workclass','occupation','native.country']]
adult_df[['age','fnlwgt','education.num','capital.gain','capital.loss','hours.per.week']] = adult_df_numeric[['age','fnlwgt','education.num','capital.gain','capital.loss','hours.per.week']]
```

#### Cible colonne "income"

```python
X = adult_df.drop(['income'], axis=1)
y = adult_df['income']
```

```python
# Remplacement <=50K et >=50K par 0 et 1
y.replace(('<=50K', '>50K'), (0, 1), inplace = True)
```

#### Split dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

#### Encodinf Categorial Values

```python
from sklearn import preprocessing
```

```python
categorical_variables = ['workclass', 'education', 'marital.status', 'occupation', 'relationship', 'race', 'sex', 'native.country']
for feature in categorical_variables:
        le = preprocessing.LabelEncoder()
        X_train[feature] = le.fit_transform(X_train[feature])
        X_test[feature] = le.transform(X_test[feature])
```

#### Scaling Variables avec StandarSacler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

cols = X_train.columns

temp_train = X_train.copy()
temp_test = X_test.copy()

X_train = pd.DataFrame(scaler.fit_transform(X_train), columns = cols)
X_test = pd.DataFrame(scaler.fit_transform(X_test), columns = cols)

print(X_train.head())
print("----------------------------------------------------------------------------")
print(X_test.head())
```

#### Test différent model
- Gradiant Boosting Classifier

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, f1_score
from sklearn.ensemble import GradientBoostingClassifier as GBC
```

```python
clf = GBC(n_estimators = 100, max_depth=5)
mod_gbc = clf.fit(X_train,y_train)
score_test = clf.score(X_test, y_test)
score_train = clf.score(X_train,y_train)
print("Score_train: ",round(score_train,3))
pred_gbc = mod_gbc.predict(X_test)
print ("Accuracy : ",round(accuracy_score(y_test, pred_gbc),3))
print ("F1_score : ", round(f1_score(y_test, pred_gbc, average = 'binary'),3))
```

#### Scv

```python
from sklearn.svm import SVC
```

```python
clf = SVC(kernel = 'rbf',max_iter = 500, probability = True)
mod_svc = clf.fit(X_train,y_train)
score_test = clf.score(X_test, y_test)
score_train = clf.score(X_train,y_train)
print("Score_train: ",round(score_train,3))
# Prediction
pred_svc = mod_svc.predict(X_test)
print ("Accuracy : ",round(accuracy_score(y_test, pred_svc),3))
print ("F1_score : ", round(f1_score(y_test, pred_svc, average = 'binary'),3))
```

#### ExtraTreesClassifier

```python
from sklearn.ensemble import ExtraTreesClassifier as ETC
```

```python
clf = ETC(n_estimators = 50, criterion="entropy", max_depth = 12)
mode_etc = clf.fit(X_train,y_train)
score_test = clf.score(X_test, y_test)
score_train = clf.score(X_train,y_train)
print("Score_train: ",round(score_train,3))
pred_etc = mode_etc.predict(X_test)
print ("Accuracy : ",round(accuracy_score(y_test, pred_etc),3))
print ("F1_score : ", round(f1_score(y_test, pred_etc, average = 'binary'),3))
```

#### AdaBoostClassifier

```python
from sklearn.ensemble import AdaBoostClassifier as ABC
```

```python
clf = ABC(n_estimators = 200, learning_rate = 1)
mode_abc = clf.fit(X_train,y_train)
score_test = clf.score(X_test, y_test)
score_train = clf.score(X_train,y_train)
print("Score_train: ",round(score_train,3))
pred_abc = mode_abc.predict(X_test)
print ("Accuracy : ",round(accuracy_score(y_test, pred_abc),3))
print ("F1_score : ", round(f1_score(y_test, pred_abc, average = 'binary'),3))
```

#### DecisionTreeClassifier 

```python
from sklearn.tree import DecisionTreeClassifier as DTC
```

```python
clf = DTC(max_depth = 7)
mode_dtc = clf.fit(X_train,y_train)
score_test = clf.score(X_test, y_test)
score_train = clf.score(X_train,y_train)
print("Score_train: ",round(score_train,3))
pred_dtc = mode_dtc.predict(X_test)
print ("Accuracy : ",round(accuracy_score(y_test, pred_dtc),3))
print ("F1_score : ", round(f1_score(y_test, pred_dtc, average = 'binary'),3))
```

#### BaggingClassifier pour le plaisir

```python
from sklearn.ensemble import BaggingClassifier as BC
```

```python
clf = BC(n_estimators = 50, n_jobs = 2)
mode_bc = clf.fit(X_train,y_train)
score_test = clf.score(X_test, y_test)
score_train = clf.score(X_train,y_train)
print("Score_train: ",round(score_train,3))
pred_bc = mode_bc.predict(X_test)
print ("Accuracy : ",round(accuracy_score(y_test, pred_bc),3))
print ("F1_score : ", round(f1_score(y_test, pred_bc, average = 'binary'),3))
```

### Classification Report

```python
for i, pred in enumerate(list_pred) :
    print ("The Confusion Matrix of : ", model_names[i])
    print (pd.DataFrame(confusion_matrix(y_test, pred)))
    print ()
```

### ROC comparaison

```python
from sklearn.metrics import roc_auc_score, roc_curve
```

```python
models = [mod_gbc, mod_svc, mode_etc, mode_abc, mode_dtc, mode_bc, model_importance, model_rfc, model_lr]
```

```python
plt.rcParams['figure.figsize'] = [10,8]
plt.style.use("bmh")

color = ['red', 'blue', 'green', 'fuchsia', 'cyan','pink', 'grey','orange','skyblue' ]
plt.title("ROC CURVE", fontsize = 15)
plt.xlabel("Specificity", fontsize = 15)
plt.ylabel("Sensitivity", fontsize = 15)
i = 1

for i, model in enumerate(models) :
    prob = model.predict_proba(X_test)
    prob_positive = prob[:,1]
    fpr, tpr, threshold = roc_curve(y_test, prob_positive)
    plt.plot(fpr, tpr, color = color[i])
    plt.gca().legend(model_names, loc = 'lower right', frameon = True)

plt.plot([0,1],[0,1], linestyle = '--', color = 'black')
plt.show()
```
